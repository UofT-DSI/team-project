Building MLR Models

Economic: ('percentage expenditure', 'Total expenditure', 'GDP', 'Status)

```python
# Define X (independent variables)
X1 = df[['percentage expenditure', 'Total expenditure', 'GDP','Status']]
# Define y (dependent variable)
y1 = df['Life expectancy']
```

```python
# Split data into training and testing sets (80% train, 20% test)

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
```

```python
# Add a constant to the model (intercept term)
X1_train = sm.add_constant(X1_train)

# Fit the MLR model
model1 = sm.OLS(y1_train, X1_train)
results = model1.fit()
summarize(results)
# Print the summary of the regression
```

Evaluate the Model1 on the Testing Data

```python
# Add a constant to the model (intercept term)
X1_test = sm.add_constant(X1_test)

# Predict using the model fitted on the training data
y1_pred = results.predict(X1_test)

# Calculate Mean Squared Error (MSE) and R-squared
mse = mean_squared_error(y1_test, y1_pred)
r2 = r2_score(y1_test, y1_pred)

print(f'MLR Mean Squared Error (MSE): {mse}')
print(f'MLR R-squared: {r2}')
```

### Plotting and Visualization

```python
# Categories

categories = {
    'Economic' : ['percentage expenditure', 'Total expenditure', 'GDP', 'Status'],
    'Infectious Diseases': ['Polio', 'Diphtheria', 'HIV/AIDS'],
    'Household income': ['Income composition of resources', 'Schooling'],
    'Lifestyle': ['Alcohol', 'BMI'],
    'Mortality': ['Adult Mortality', 'infant deaths', 'thinness  1-19 years'],
    }
```

```python
# Create a common plotting

def plot_relationships(df, category_name, variables):
    plt.figure(figsize=(14,8))
    for variable in variables:
        if variable != 'Status':
            sns.scatterplot(x=variable, y='Life expectancy', data=df, label=variable)
    plt.title(f'{category_name}: the relationship between variables and Life expectancy')
    plt.xlabel('variable')
    plt.ylabel('Life expectancy')
    plt.legend()
    plt.show()
```

```python
#Economic
plot_relationships(df,'economics', [var for var in categories['Economic'] if var !="Status"])

#Map numerical values to lables
df ['Status'] = df ['Status'].map({1: 'Developing',0: 'Developed'})

#plotting
plt.figure(figsize=(14,8))
sns.boxplot(x='Status',y='Life expectancy', data=df,hue = "Status", hue_order= ['Developing','Developed'])
plt.title('Life expectancy by status (Developed vs Developing)')
plt.xlabel('Status')
plt.ylabel('Life expantancy')
plt.show()
```

Function to perform EDA and generate plots

```python
def perform_eda(df):
    plt.figure(figsize=(18,15))
    total_plots = sum(len(vars) for vars in categories.values())

    #Iterate through categories and variables
    plot_index =1
    for category, variables in categories.items():
        for var in variables:
            plt.subplot(total_plots, 2, plot_index)
            sns.scatterplot(x=var, y='Life expectancy', data=df)
            plt.title(f'{category}: {var} vs Life expectancy')
            plot_index +=1

            plt.subplot(total_plots, 2, plot_index)
            sns.histplot(df[var].dropna(), kde=True)
            plt.title(f'{category}: Distribution of {var}')
            plot_index +=1

    plt.subplots_adjust(top=2.3, bottom=0.01, left=0.1, right=0.9, hspace=0.35, wspace = 0.25)
    plt.show()

# Perform EDA
perform_eda(df)   
```
