```python
for predictor in df.columns.tolist():
    if (predictor != 'salary') & (predictor != 'job_title'):
      model = sm.OLS.from_formula('salary ~ ' + predictor, data=df)
      result = model.fit()
      print('======================= ' + predictor + ' ==================================================')  
      print(result.summary())
```
