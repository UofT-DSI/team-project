```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
from scipy.stats import ttest_ind

import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv('../data/raw/Walmart_Store_sales.csv')


data.head()

data.isnull().sum()

data.describe()
```

Check relationship between Weekly_Sales and Holiday_Flag

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Select the feature and the target variable
X = data[['Holiday_Flag']]
y = data['Weekly_Sales']

# Add a constant term for the intercept
X = sm.add_constant(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Fit the linear regression model
model = sm.OLS(y_train, X_train).fit()

# Print the model summary
print(model.summary())
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the mean squared error and R^2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R^2 Score: {r2:.2f}')
```

```python
# Plot actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.xlabel('Actual Weekly Sales')
plt.ylabel('Predicted Weekly Sales')
plt.title('Actual vs Predicted Weekly Sales')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2)
plt.show()
```

Results: Overall, the results indicate that Holiday_Flag does not have a significant impact on Weekly_Sales based on this linear regression model.

Check relationship between Weekly_Sales and Fuel_Price

```python
# # Extracting the variables
X = data[['Fuel_Price']]
y = data['Weekly_Sales']

X = sm.add_constant(X)  # Add a constant term to the predictor

# Regression model
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X['Fuel_Price'], y, alpha=0.5, label='Data')

# Plotting the regression line
plt.plot(X['Fuel_Price'], results.predict(), color='red', label='Regression Line')

plt.title('Weekly Sales vs Fuel Price')
plt.xlabel('Fuel Price')
plt.ylabel('Weekly Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("R-squared:", results.rsquared)
```

The model shows that fuels prices do not affect weekly sales.
